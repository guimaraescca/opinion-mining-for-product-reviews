```python
import pandas as pd
import rdflib
import rdflib.plugins.sparql as sparql
from rdflib.namespace import RDF, RDFS, OWL
```

```python
g = rdflib.Graph()
g.load('../data/external/ontologies/smartphone_aspects.owl')

search_term = 'som'
search_term = search_term.lower()

query_str = """
        SELECT DISTINCT ?individualLabel ?classLabel 
        WHERE {
            ?y rdf:type owl:Class . 
            ?y rdfs:label ?classLabel . 
            OPTIONAL{
                ?x rdf:type ?y .
                ?x rdfs:label ?individualLabel .
            }
            FILTER( REGEX(str(?individualLabel), "^%s([_]|$)|([_]|^)%s$", "i") ||
                    REGEX(str(?classLabel), "^%s([_]|$)|([_]|^)%s$", "i")) 
        } 
        ORDER BY ASC(?classLabel)""" % (search_term, search_term, search_term, search_term)

query = sparql.prepareQuery(query_str, initNs = {'rdf': RDF, 'rdfs': RDFS , 'owl': OWL})
query_result = g.query(query, DEBUG = True)

data = []
for row in query_result:
    data_row = []
    for x in row:
        if x is not None: 
            x = x.toPython()
        data_row.append(x)
    data.append(data_row)

df = pd.DataFrame(data, columns=['individualLabel', 'classLabel'])

try:
    result_class = df.loc[0, 'classLabel']
except KeyError:
    result_class = None

print(f'Resulted cluster: {result_class}')
display(df)
```
